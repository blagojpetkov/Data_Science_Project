```python
import numpy as np
import pandas as pd
from urllib.parse import urlparse

import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
# Setting up 'requests' to make HTTPS requests properly takes some extra steps... we'll skip them for now.
requests.packages.urllib3.disable_warnings()
# Import all modules that are required
%matplotlib inline
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
def generate_week_strings(start_year, start_week, end_year, end_week):
    week_strings = []

    for year in range(start_year, end_year + 1):
        start = start_week if year == start_year else 1
        end = end_week if year == end_year else 53

        for week in range(start, end + 1):
          #TODO CHANGE LATER range(1, 11)
          for num in range(1, 3):
            #TODO CHANGE LATER range(1, 11)
            week_string = f"week/{year}/{week}/{num}"
            week_strings.append(week_string)

    return week_strings

start_year = 2023
start_week = 19
end_year = 2023
end_week = 20

week_strings = generate_week_strings(start_year, start_week, end_year, end_week)
# for week_string in week_strings:
#     print(week_string)
```

```python
len(week_strings)
```

```python
list_of_articles = []
list_of_urls = []
for week_string in week_strings:
  week_url = "https://time.mk/" + week_string
  response = requests.get(week_url)
  response.encoding = 'utf-8'
  soup = BeautifulSoup(response.text, "html.parser")
  list_of_articles.extend(soup.find_all(class_="cluster"))
for article in list_of_articles:
  first_anchor_tag = article.select_one('a')
  list_of_urls.append("https://time.mk/" + first_anchor_tag.get("href"))
```

```python
print(list_of_urls)
```

```python
len(list_of_urls)
```

```python
real_urls_list = []
successfully_accessed_sites = []
successfully_accessed_hostnames = []
unsuccessfully_accessed_sites = []
for time_url in list_of_urls[0:50]:
  response = requests.get(time_url)
  real_url = response.url
  real_urls_list.append(real_url)
  if response.status_code==200:
    successfully_accessed_sites.append(real_url)
    successfully_accessed_hostnames.append(urlparse(real_url).netloc)
  else:
    unsuccessfully_accessed_sites.append(real_url)
```

```python
import transformers
import nlp
import torch
from transformers import pipeline
from transformers import MarianMTModel, MarianTokenizer

# Load the pre-trained model and tokenizer
model_name = "Helsinki-NLP/opus-mt-mk-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)
```

```python
import re

url_text_list = list(url_text_map.items())
translated_texts = {}
for url_text in url_text_list[:1]:

    # Define the input Macedonian text
    
    input_url = url_text[0]
    input_text = url_text[1]

    # Split text into sentences
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', input_text)

    # Translate each sentence separately
    translated_sentences = []
    for sentence in sentences:
        input_ids = tokenizer.encode(sentence, return_tensors="pt")
        with torch.no_grad():
            translated_ids = model.generate(input_ids)
        translated_sentence = tokenizer.decode(translated_ids[0], skip_special_tokens=True)
        translated_sentences.append(translated_sentence)

    # Combine translated sentences
    translated_text = " ".join(translated_sentences)
    translated_texts[input_url] = translated_text
```

```python
import spacy

def sentiment(text):
  if text in ["bad", "terrible", "awful"]:
    return "negative"
  elif text in ["good", "great", "wonderful"]:
    return "positive"
  else:
    return "neutral"


nlp = spacy.load('en_core_web_sm')

text = "Canada, with its breathtaking landscapes and diverse cities, stands as a model of inclusivity and respect. Its warm-hearted citizens and rich cultural tapestry create a welcoming atmosphere, while its commitment to arts, culture, and sustainability further elevates its global reputation. A nation that embraces both natural beauty and progressive values, Canada's influence resonates far beyond its borders."

doc = nlp(text)

for entity in doc.ents:
  print(entity.text, entity.label_, sentiment(entity.text))
```

```python
import spacy
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load the SpaCy NER model and VADER sentiment analyzer
nlp = spacy.load("en_core_web_sm")
analyzer = SentimentIntensityAnalyzer()

text="Canada is horrific, murder, kill"
# Process the text using SpaCy
doc = nlp(text)

# Initialize a dictionary to store named entities and sentiments
entity_sentiments = {}

# Extract named entities and analyze sentiment for each
for entity in doc.ents:
    entity_text = entity.text
    sentiment = analyzer.polarity_scores(entity_text)["compound"]
    print(sentiment)
    # Categorize sentiment
    if sentiment >= 0.05:
        sentiment_label = "Positive sentiment"
    elif sentiment <= -0.05:
        sentiment_label = "Negative sentiment"
    else:
        sentiment_label = "Neutral sentiment"
    
    entity_sentiments[entity_text] = sentiment_label

# Print the results
for entity, sentiment in entity_sentiments.items():
    print(f"{entity}: {sentiment}")
```

```python
import spacy
from textblob import TextBlob

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "The new smartphone has great features, but the customer service was terrible."

# Process text using spaCy
doc = nlp(text)

# Initialize dictionary to store entity sentiments
entity_sentiments = {}

# Calculate overall sentiment of the sentence
overall_sentiment = TextBlob(text).sentiment.polarity

# Iterate through entities in the text

print(doc.ents)
for entity in doc.ents:
    print(entity.sentiment)
    entity_text = entity.text
    entity_sentiment = overall_sentiment + entity.sentiment
    
    entity_sentiments[entity_text] = entity_sentiment

print("Entity Sentiments:")
for entity, sentiment in entity_sentiments.items():
    print(f"{entity}: {sentiment:.2f}")
```
